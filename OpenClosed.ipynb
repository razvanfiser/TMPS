```python
import numpy as np
import matplotlib.pyplot as plt
from abc import ABC, abstractmethod
```

## Open-closed Principle

```python
# class World inherits from the ABC (or Abstract Base Class) which is the implementation of an abstract class in Python
class WorldBuilder(ABC):
    @abstractmethod
    def build_world(self, n_samples):
        pass

# classes below inherit from abstractclass WorldBuilder and produce variously-shaped worlds with tiny organisms
# code is open to extension, because if other shapes need be added, w
class DiscWorld(WorldBuilder):
    def __init__(self):
        pass
        
    def build_world(self, n_samples):
        radii = np.random.uniform(size=n_samples)
        angles = np.random.uniform(0, 2*3.14, size=n_samples)
        plt.scatter(radii * np.cos(angles), radii * np.sin(angles))
        
class SquareWorld(WorldBuilder):
    def __init__(self):
        pass
    
    def build_world(self, n_samples):
        plt.scatter(np.random.uniform(size=n_samples), np.random.uniform(size=n_samples))
```

```python
DiscWorld().build_world(10000)
```

```python
SquareWorld().build_world(10000)
```
